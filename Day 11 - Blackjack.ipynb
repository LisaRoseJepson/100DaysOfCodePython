```python
def draw_card(player):
    card = deck.pop(random.randint(0, len(deck)-1))
    if player == "user":
        user_cards.append(card)
    else:
        computer_cards.append(card)
```

```python
def setup_game():
    global deck
    global user_cards
    global computer_cards
    
    deck = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10] * 4
    
    user_cards = []
    computer_cards = []
    
    draw_card("user")
    draw_card("user")
    draw_card("computer")
    draw_card("computer")
    
    print(logo)
```

```python
def display_cards(player):
    if player == "user":
        print(f"Player's hand: {user_cards}")
    else:
        hide_first = computer_cards.copy()
        hide_first[0] = "X"
        print(f"Computer's hand: {hide_first}")
```

```python
def bust2(deck):
    
    total = hand_total(deck)
    
    if total <= 21: # if now equal to or under under 21, not bust
        return False
    else:
        return True # if still over 21, bust   
```

```python
def hand_total(deck):
    
    total = 0
    ace_count = 0
    
    for card in deck:
        total += card
        if card == 11:
            ace_count += 1
        
    while ace_count > 0 and total > 21: # reduce the total by 10 for each ace, until total under 21 or no aces left
        ace_count -= 1
        total -= 10
        
    return total
```

```python
play = True

while play:
    
    setup_game()
    display_cards("user")
    display_cards("computer")

    play_on = True
    player_hold = False
    computer_hold = False

    while play_on:
        if not player_hold and play_on:

            draw = input("Would you like to draw a card? Y/N: ").upper()

            if draw == "N":
                player_hold = True
            else:
                draw_card("user")
                display_cards("user")

            play_on = not bust2(user_cards)

        if not computer_hold and play_on: 
            computer_total = hand_total(computer_cards)

            user_total = hand_total(user_cards)

            if computer_total < 17 or (user_total > computer_total):
                draw_card("computer")
            else:
                computer_hold = True

            if bust2(computer_cards):
                play_on = False
                print(computer_cards)
            else:
                display_cards("computer")

        if player_hold and computer_hold:
            play_on = False

    print(user_cards)
    print(computer_cards)

    if bust2(user_cards):
        print("Bust! You lose...")
    elif bust2(computer_cards):
        print("House bust... You win!")
    else:
        user_total = hand_total(user_cards)
        computer_total = hand_total(computer_cards)

        if user_total > computer_total:
            print(f"Player {user_total}, house {computer_total}: You win!")
        elif computer_total > user_total:
            print(f"House {computer_total}, player {user_total}: House wins!")
        else:
            print(f"Player {user_total}, house {computer_total}: Game tied!")  
            
    play_again = input("Would you like to play again? Y/N: ").upper()
    
    if play_again == "N":
        play = False
```
